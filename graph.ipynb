```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from pylab import rcParams
import networkx as nx
```

```python
%matplotlib inline
rcParams['figure.figsize']=7,4
sb.set_style('whitegrid')
```

```python
dg=nx.gn_graph(7,seed=25)
for line in nx.generate_edgelist(dg,data=False):
    print(line)
```

```python
dg.node[0]['name']='alice'
dg.node[1]['name']='bob'
dg.node[2]['name']='piyush'
dg.node[3]['name']='lavish'
dg.node[4]['name']='mota'
dg.node[5]['name']='sarkar'
dg.node[6]['name']='bpsj'
```

```python
dg.add_nodes_from([(0,{'age':25}),(1,{'age':21}),(2,{'age':20}),(3,{'age':32}),(4,{'age':26}),(5,{'age':29}),(6,{'age':16})])
```

```python
dg.add_nodes_from([(0,{'gender':'m'}),(1,{'gender':'f'}),(2,{'gender':'f'}),(3,{'gender':'m'}),(4,{'gender':'m'}),(5,{'gender':'f'}),(6,{'age':'m'})])
```

```python
nx.draw_circular(dg,node_color='bisque',with_labels=True)
```

```python
labeldict={0:'alice',1:'bob',2:'piyush',3:'lavish',4:'mota',5:'sarkar',6:'bpsj'}
nx.draw_circular(dg,labels=labeldict,node_color='bisque',with_labels=True)
```

```python
d=dg.to_undirected()
```

```python
nx.draw_spectral(dg,labels=labeldict,node_color='bisque',with_labels=True)
```
